# Creating visualization for the Wine Dataset

```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import dataset
wine_data = pd.read_csv('../resources/winequalityN.csv')
```

```python
# view first 5 entries of wine_data
wine_data.head()
```

### get summary statistics on the dataframe

```python
wine_data.describe(include='all')
```

### find total number of red and white wines in the dataset and create bar graph to show difference

```python
# group by type of wine
wine_grouped = wine_data.groupby(["type"]).count()["quality"]

# get keys for wine_grouped
wine_types = wine_grouped.keys()

# get values for wine_grouped
wine_type_counts = wine_grouped.values

ax = plt.subplots()
plt.bar(wine_types, wine_type_counts)
plt.title('count of bottles in dataset per wine type')
plt.xlabel('type of wine')
plt.ylabel('number of bottles')
plt.show()
```
